# Read-Image

#### Repositório para testes da biblioteca OpenCV,  o projeto destinasse a identificar elementos através imagens

```python
#importa todas as bibliotecas usadas
import matplotlib.pyplot as plt
import numpy as np
import imutils
import cv2 as cv
```

Lendo imagem contida no dataSet

```python
img = cv.imread('dataset/imgs/preto.jpg')
```

Vamos apresentar a imagem coletada

```python
plt.imshow(img)
plt.show()
```

selecionando ponto preto no meio da imagem

```python
point = img[ 80:145,90:155 ]
```

O ponto preto no meio da imagem foi selecionado bem próximo de sua origem, por isso ao imprimir o mesmo estará o mais próximo possível da origem (x=0:0)

```python
plt.imshow(point)
plt.show()
```

Agora adicionaremos o ponto coletado em outro campo da imagem original

```python
img[45:110,64:129] = point #ps: as dimensões da nova inserção devem respeitar as dimensões da imagem a ser inserida
```

```python
plt.imshow(img)
plt.show()
```

Quais outras operações podemos efetuar na imagem?

Rotacioar:

```python
for angle in np.arange(0, 360, 15):
    rotated = imutils.rotate(img, angle)
    plt.imshow(rotated)
    plt.show()
```
